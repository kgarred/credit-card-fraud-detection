```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the credit card fraud detection dataset

filename = os.path.join(os.path.dirname(__name__), "../data/creditcard.csv")
creditcard_data = pd.read_csv(filename, delimiter=',', low_memory=False)
print(creditcard_data.info())
creditcard_data.head()
```

```python
# Check the null values in the dataset

creditcard_data.isnull().sum()
```

```python
# Describe the creditcard dataset

creditcard_data.describe()
```

```python
# Check the balance with the distribution of the target variable 'Class'

creditcard_data.Class.value_counts()
```

```python
# Plot the distribution of the target variable 'Class'

plt.figure(figsize=(6,4))
creditcard_data['Class'].value_counts().plot(kind='bar')
plt.title('Distribution of Class Variable')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.show()
```
