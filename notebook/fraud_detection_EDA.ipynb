```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the credit card fraud detection dataset

filename = os.path.join(os.path.dirname(__name__), "../data/creditcard.csv")
creditcard_data = pd.read_csv(filename, delimiter=',', low_memory=False)
print(creditcard_data.info())
creditcard_data.head()
```

```python
# Check the null values in the dataset

creditcard_data.isnull().sum()
```

```python
# Describe the creditcard dataset

creditcard_data.describe()
```

```python
# Check the balance with the distribution of the target variable 'Class'

creditcard_data.Class.value_counts()
```

```python
balance_data = {'class': ['fault','non-fault'], 'count': creditcard_data['Class'].value_counts().values.tolist()}
balance_data
```

```python
# Plot the distribution of the target variable 'Class'

balance_data = {'class': ['fault','non-fault'], 'count': creditcard_data['Class'].value_counts().values.tolist()}
plt.figure(figsize=(6,4))
plt.bar(balance_data['class'], balance_data['count'], color='cornflowerblue', log=False)
plt.title('Distribution of Class Variable')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the distribution of the target variable 'Class'

balance_data = {'class': ['fault','non-fault'], 'count': creditcard_data['Class'].value_counts().values.tolist()}
plt.figure(figsize=(6,4))
plt.bar(balance_data['class'], balance_data['count'], color='cornflowerblue', log=True)
plt.title('Log Distribution of Class Variable')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.show()
```

```python
# Amount distribution by Class

sns.boxplot(x='Class', y='Amount', data=creditcard_data)
plt.title('Amount Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Amount')
plt.show()
```

```python
# Time patterns (fraud occurrences by hour)

creditcard_data['hour'] = (creditcard_data['Time'] / 3600).astype(int) % 24
```

```python
fraud_by_hour = creditcard_data[creditcard_data['Class'] == 1]['hour'].value_counts().sort_index()
normal_by_hour = creditcard_data[creditcard_data['Class'] == 0]['hour'].value_counts().sort_index()
```

```python
plt.figure(figsize=(14,6))
sns.barplot(x=fraud_by_hour.index, y=fraud_by_hour.values)
plt.title("Fraud Transactions by Hour of Day")
plt.xlabel("Hour of Day (0–23)")
plt.ylabel("Number of Fraud Transactions")
plt.show()
```

```python
# Correlation heatmap

sns.heatmap(creditcard_data.corr(), cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
